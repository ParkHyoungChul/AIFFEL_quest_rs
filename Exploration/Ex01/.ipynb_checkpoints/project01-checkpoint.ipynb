### 당뇨병 데이터셋 예시로 보는 다중 선형 회귀

이 데이터셋에는 442명의 당뇨병 환자에 대한 10개의 기본 변수(연령, 성별, 체질량 지수, 평균 혈압, 혈청 측정치 등)가 포함되어 있습니다. 또한, 각 환자에 대해 1년 후 당뇨병 진행 정도를 나타내는 수치가 목표 변수로 주어집니다.

**샘플 개수**:
442개의 샘플(환자 데이터)이 존재합니다.

**속성 개수**:
10개의 속성(변수)가 예측에 사용됩니다. 이 변수들은 모두 숫자형입니다.

**목표값**:
목표값은 11번째 열로, 각 환자의 1년 후 당뇨병 진행 정도를 나타내는 정량적 지표입니다.


**속성(변수) 정보**:
- age: 나이 (연도 기준)
- sex: 성별
- bmi: 체질량 지수 (Body Mass Index)
- bp: 평균 혈압 (Blood Pressure)
- s1: 총 혈청 콜레스테롤 (Total Serum Cholesterol, tc)
- s2: 저밀도 지질 단백 (Low-Density Lipoproteins, ldl)
- s3: 고밀도 지질 단백 (High-Density Lipoproteins, hdl)
- s4: 총 콜레스테롤/HDL 비율 (Total Cholesterol / HDL, tch)
- s5: 아마도 혈청 중성지방 수준의 로그 값 (Log of Serum Triglycerides Level, ltg)
- s6: 혈당 수치 (Blood Sugar Level, glu)

```python
# 데이터 셋 로드 
from sklearn.datasets import load_diabetes 

# sklearn 자체 제공 데이터셋 diabetes
# 함수 형태로 불러올 경우 ndarray 형태로 가져온다 (sklearn.utils.Bunch ? )
diabetes = load_diabetes()
```

```python
diabetes["data"].shape
```

```python
# 데이터 가져오기 -
# data 행렬과 타겟 y 벡터 준비 

import numpy as np

df_X = np.array(diabetes["data"])

# age,sex s1,s2 열 제거 - 0,1,4,5

df_X = np.delete(df_X, [0, 1, 4, 5], axis=1)
df_y = np.array(diabetes["target"])
```

```python
df_X.shape
```

```python
# 훈련용 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=33)
```

```python
# 모델 준비하기 
# 가중치 초기값 설정
W = np.random.rand(6)
b = np.random.rand()

# 모델 함수 구현
def model(X,W,b):
    prediction = 0
    for i in range(0, len(W)):
        prediction += X[:,i]*W[i]
    prediction += b
    return prediction      
    
```

```python
# 손실함수 정의 (MSE)
def loss(y_pred, y):
    result = np.mean((y_pred - y)**2)
    return result
```

```python
# gradient 함수 정의
def gradient(X,y,W,b):
    
    N = len(y)
    y_pred = model(X,W,b)
    
    dW = 1/N * 2 * X.T.dot(y_pred - y)
    db = 2 * (y_pred - y).mean()
    return dW, db    
```

```python
# 학습률 설정

learning_rate = 0.2
```

### 학습률 history

epoch는 1000으로 고정
TEST LOSS
- 0.01 = 5000얼마
- 0.1 = 3011.096125478127
- 0.5 = 2710.979622612038
- 0.6 = 2728.401677356017
- 1 = 31340.778031164864

##### 1로 하니까 갑자기 뒤에서 LOSS가 심해짐



### 컬럼을 제외하고 나서 진행한 결과 

rate = 0.5  
loss = 2597.3916978618195

rate = 0.3  
loss = 2696.9610165734607

```python
epoch = 1000
losses = []
for i in range(1,epoch):
    dW, db = gradient(X_train,y_train,W,b)
    W -= learning_rate*dW
    b -= learning_rate*db
    y_pred = model(X_train,W,b)
    L = loss(y_pred, y_train)
    losses.append(L)
    if i%10 == 0:
        print('Iteration %d : Loss %0.4f' % (i, L))
```

```python
# 학습 그래프 살펴보기
import matplotlib.pyplot as plt
plt.plot(losses)
plt.show()
```

```python
# 테스트 성능 확인
prediction = model(X_test, W, b)
mse = loss(prediction, y_test)
mse
```
