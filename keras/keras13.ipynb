```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.model_selection import train_test_split
import os
```

```python
def get_mnist_model():
    inputs = keras.Input(shape=(28 * 28,))
    features = layers.Dense(512, activation="relu")(inputs)
    features = layers.Dropout(0.5)(features)
    outputs = layers.Dense(10, activation="softmax")(features)
    model = keras.Model(inputs, outputs)
    return model
```

```python
(images, labels), (test_images, test_labels) = keras.datasets.mnist.load_data()
images = images.reshape((60000, 28 * 28)).astype("float32") / 255
test_images = test_images.reshape((10000, 28 * 28)).astype("float32") / 255
train_images, val_images = images[10000:], images[:10000]
train_labels, val_labels = labels[10000:], labels[:10000]
```

```python
model = get_mnist_model()
model.summary()
```

```python
model.compile(optimizer="rmsprop",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
history = model.fit(train_images, train_labels,
          epochs=10,
          validation_data=(val_images, val_labels))
```

```python
test_metrics = model.evaluate(test_images, test_labels)
```

```python
def remap_labels(labels):
    unique_labels = np.unique(labels)
    label_mapping = {label: idx for idx, label in enumerate(unique_labels)}
    return np.array([label_mapping[label] for label in labels])

# y_train과 y_test에 대해 레이블 재매핑
y_train_mapped = remap_labels(y_train)
y_train0_mapped = remap_labels(y_train0)
y_val_mapped = remap_labels(y_val)
y_test_mapped = remap_labels(y_test)
```

```python
class LossHistory(keras.callbacks.Callback):
    def on_train_begin(self, logs):
        self.per_batch_losses = []

    def on_batch_end(self, batch, logs):
        self.per_batch_losses.append(logs.get("loss"))

    def on_epoch_end(self, epoch, logs):
        plt.clf()
        plt.plot(range(len(self.per_batch_losses)), self.per_batch_losses,
                 label="Training loss for each batch")
        plt.xlabel(f"Batch (epoch {epoch})")
        plt.ylabel("Loss")
        plt.legend()
        plt.savefig(f"plot_at_epoch_{epoch}")
        self.per_batch_losses = []
```

```python
model.fit(X_train0, y_train0_mapped,
          epochs=50,
          callbacks=[LossHistory()],
          validation_data=(X_val, y_val_mapped))
```

```python
import keras_tuner
import numpy as np
```

```python
wandb.init(
    project = "3-wandb-project",

    config = {
        "epochs" : 20,
        "dataset" : 'wine_q',
    }
)
```

```python
from wandb.integration.keras import WandbCallback

WandbCallback(
    monitor='val_loss',  # 모니터링할 메트릭
    save_weights_only=True  # 가중치만 저장
)
```

```python
def build_wine_model(hp):
    inputs = keras.Input(shape=(11,))
    # 모델을 두 가지 타입으로 사용해봅니다
    x = inputs

    for i in range(hp.Int("mlp_layers", 1, 3)):  # 레이어 개수

        # 레이어 이름을 다르게 만들기
        x = layers.Dense(
            units=hp.Int(f"units_{i}", 8, 64, step=8),   # 유닛
            activation="relu",
        )(x)

    # 드랍아웃을 사용하는 경우와 사용하지 않는 경우를 구분  
    if hp.Boolean("dropout"):
        x = layers.Dropout(0.2)(x)

    outputs = layers.Dense(units=7, activation="softmax")(x)
    model = keras.Model(inputs=inputs, outputs=outputs)

    model.compile(
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
        optimizer="adam",
    )
    return model
```

```python
hp = keras_tuner.HyperParameters()
# 설정한 hp 로 모델을 만듭니다 -> 전체 집합 중 부분집합이 생성됩니다
model = build_wine_model(hp)
model.summary()
```

```python
tuner = keras_tuner.RandomSearch(
    build_wine_model,
    max_trials=20,
    overwrite=True,
    objective="val_accuracy",
    directory="/tmp/tb", # 디렉토리를 설정해주었습니다
)
```

```python
tuner.search(
    X_train,
    y_train_mapped,
    validation_split=0.2,
    epochs=10,
    callbacks=[WandbCallback()], # 위 설정한 디렉토리에 callback으로 로그가 기록됩니다
)
```
